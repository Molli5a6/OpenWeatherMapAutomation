```python

import requests

API_KEY = "b6907d289e10d714a6e88b30761fae22"
BASE_URL = "https://samples.openweathermap.org/data/2.5/forecast/hourly"

def get_weather_data(city):
    url = f"{BASE_URL}?q={city}&appid={API_KEY}"
    response = requests.get(url)
    if response.status_code == 200:
        return response.json()
    else:
        print("Error fetching weather data. Please check your input or try again later.")
        return None

def get_temperature_on_date(weather_data, date):
    for forecast in weather_data['list']:
        if forecast['dt_txt'][:10] == date:
            return forecast['main']['temp']
    return None

def get_wind_speed_on_date(weather_data, date):
    for forecast in weather_data['list']:
        if forecast['dt_txt'][:10] == date:
            return forecast['wind']['speed']
    return None

def get_pressure_on_date(weather_data, date):
    for forecast in weather_data['list']:
        if forecast['dt_txt'][:10] == date:
            return forecast['main']['pressure']
    return None

def main():
    city = input("Enter the city (e.g., London): ")
    weather_data = get_weather_data(city)

    if weather_data:
        while True:
            print("\nOptions:")
            print("1. Get weather")
            print("2. Get Wind Speed")
            print("3. Get Pressure")
            print("0. Exit")

            option = int(input("Select an option: "))

            if option == 1:
                date = input("Enter the date (YYYY-MM-DD): ")
                temperature = get_temperature_on_date(weather_data, date)
                if temperature is not None:
                    print(f"Temperature on {date}: {temperature} K")
                else:
                    print("Invalid date or date not found in the forecast.")
            elif option == 2:
                date = input("Enter the date (YYYY-MM-DD): ")
                wind_speed = get_wind_speed_on_date(weather_data, date)
                if wind_speed is not None:
                    print(f"Wind Speed on {date}: {wind_speed} m/s")
                else:
                    print("Invalid date or date not found in the forecast.")
            elif option == 3:
                date = input("Enter the date (YYYY-MM-DD): ")
                pressure = get_pressure_on_date(weather_data, date)
                if pressure is not None:
                    print(f"Pressure on {date}: {pressure} hPa")
                else:
                    print("Invalid date or date not found in the forecast.")
            elif option == 0:
                print("Exiting the program.")
                break
            else:
                print("Invalid option. Please try again.")

if __name__ == "__main__":
    main()
```
